```python
import math
a,b,c=map(int,input("Nhập lần lượt 3 cạnh của tam giác ").split())
s=(a+b+c)/2

if a==b==c :
    print("Đây là tam giác đều")
    print("Diện tích là ",a**2*(math.sqrt(3))*0.25)

elif a==b or b==c or c==a :
    print("Đây là tam giác cân")
    if a==b :
        print("Diện tích là ",0.5*c*(math.sqrt(b**2-(c/2)**2)))
    elif a==c :
        print("Diện tích là ",0.5*b*(math.sqrt(c**2-(b/2)**2)))
    else:
        print("Diện tích là ",0.5*a*(math.sqrt(b**2-(a/2)**2)))
    

elif a**2 +b**2 == c**2 or a**2 +c**2 == b**2 or c**2 +b**2 == a**2 :
    print("Đây là tam giác vuông ")
    if a**2 +b**2 == c**2 :
        print("Diện tích tam giác là :",0.5*a*b)
    elif a**2 +c**2 == b**2 :
        print("Diện tích tam giác là :",0.5*a*c)
    else :
        print("Diện tích tam giác là :",0.5*b*c)
elif a<b+c and b<a+c and c<a+b:
    print("Đây là tam giác thường")
    print("Diện tích tam giác là :",math.sqrt(s * (s - a) * (s - b) * (s - c)))
else:
    print("Đây không là 3 cạnh của tam giác")
```

```python
a,b,m,c,d,n=map(int,input("Nhập 6 số tự nhiên bất kì a,b,m,c,d,n theo thứ tự").split())
delta=a*d-b*c
deltax=m*d-n*b
deltay=a*n-c*m
if a*d-b*c != 0 :

    print("Nghiệm của hệ phương trình tuyến tính là ")
    print("x1=",deltax/delta)
    print("x2=",deltay/delta)
else :
    if a*d-b*c ==0 and m*d-n*b ==0 and a*n-m*c ==0 :
        print("Hệ có vô số nghiệm ")
    else :
        print("Hệ vô nghiệm")
```

Bài 10

```python
import math
xi,yi=map(float,input("Hãy nhập hoành độ  và tung độ tâm I ").split())
r=int(input("Hãy nhập bán kính hình tròn "))
x,y=map(float,input("Hãy nhập hoành độ  và tung độ điểm A ").split())
khoangcach = math.sqrt((xi-x)**2+(yi-y)**2)
if khoangcach == r :
    print("Điểm A thuộc đường tròn")
elif khoangcach <r :
    print("Điểm A nằm trong đường tròn")
else :
    print("Điểm A nằm ngoài đường tròn")
```
